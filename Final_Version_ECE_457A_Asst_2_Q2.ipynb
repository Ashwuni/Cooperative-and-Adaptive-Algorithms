```python
#DFS
```

```python
class Node:
    def __init__(self, point, nodeParent):
        self.point = point
        self.nodeParent = nodeParent        
    
    def __lt__(self, other_node):
        return True

    def get_children(self, maze):
        children = []
        x = self.point[0]
        y = self.point[1]
        if x - 1 >= 0 and not maze[y][x-1]:
            children.append((x-1, y))
        if x + 1 < len(maze[0]) and not maze[y][x+1]:
            children.append((x+1, y))
        if y - 1 >= 0 and not maze[y-1][x]:
            children.append((x, y-1))
        if y + 1 < len(maze) and not maze[y+1][x]:
            children.append((x, y+1))
        return children
```

```python
def dfs(maze, start, end):
    open_list = [start] 
    root = Node(start, None)
    open_list.append(root)
    closed_list = [] 
    node = root
    visited = 0
    while True:
      while node.point in closed_list:
          visited = visited +1
          node = open_list.pop()
      if node.point == end:
          visited = visited +1
          return node, visited
      closed_list.append(node.point)
      children = node.get_children(maze)
      children.reverse()
      for child in children:
          if not child in closed_list:
              open_list.append(Node(child, node))
```

```python
#c1

answer = dfs(maze, (2, 13),(2, 3))
node = answer[0]
print("The path is as follows: ")
cost = -1
while node != None:
    print(node.point)
    cost = cost + 1
    node = node.nodeParent
print("Cost of the path: ", cost)
print("The number of nodes explored is as follows: ", answer[1])
```

```python
class Node2:
    def __init__(self, point, nodeParent):
        self.point = point
        self.nodeParent = nodeParent
    
    def __lt__(self, other_node):
        return True

    def get_children(self, maze):
        children = []
        x = self.point[0]
        y = self.point[1]
        if y + 1 < len(maze) and not maze[y+1][x]:
            children.append((x, y+1))
        if x + 1 < len(maze[0]) and not maze[y][x+1]:
            children.append((x+1, y))
        if y - 1 >= 0 and not maze[y-1][x]:
            children.append((x, y-1))
        if x - 1 >= 0 and not maze[y][x-1]:
            children.append((x-1, y))    
        return children
```

```python
def bfs(maze, start, end):
    root = Node2(start, None)
    open_list = [root]
    open_list.append(root)
    closed_list = []
    node = root
    visited = 0
    while True:
      while node.point in closed_list:
          node = open_list.pop(0)
          visited = visited +1
      if node.point == end:
          visited = visited +1
          return node, visited
      closed_list.append(node.point)
      for child in node.get_children(maze):
          if not child in closed_list:
              open_list.append(Node2(child, node))
```

```python
class Node3:
    def __init__(self, point, nodeParent, distance, end):
        self.point = point
        self.nodeParent = nodeParent
        self.distance = distance 
        self.heuristic = abs(point[0]-end[0]) + abs(point[1]-end[1])
    
    def __lt__(self, other_node):
        return True

    def get_children(self, maze):
        children = []
        x = self.point[0]
        y = self.point[1]
        if y + 1 < len(maze) and not maze[y+1][x]:
            children.append((x, y+1))
        if x + 1 < len(maze[0]) and not maze[y][x+1]:
            children.append((x+1, y))
        if y - 1 >= 0 and not maze[y-1][x]:
            children.append((x, y-1))
        if x - 1 >= 0 and not maze[y][x-1]:
            children.append((x-1, y))    
        return children
```

```python
def astar(maze, start, end):
    root = Node3(start, None, 0, end)
    open_list = [root]
    open_list.append(root)
    closed_list = []
    node = root
    visited = 0
    while True:
      while node.point in closed_list:
          open_list.sort(key=lambda x: x.distance + x.heuristic)
          node = open_list.pop()
          visited = visited+1
      if node.point == end:
          visited = visited +1
          return node, visited
      closed_list.append(node.point)
      for child in node.get_children(maze):
          if not child in closed_list:
              open_list.append(Node3(child, node, node.distance + 1, end))
```

```python
#c1

answer = astar(maze, (2, 13),(2, 3))
node = answer[0]
print("The path is as follows: ")
cost = -1
while node != None:
    print(node.point)
    cost = cost + 1
    node = node.nodeParent
print("Cost of the path: ", cost)
print("The number of nodes explored is as follows: ", answer[1])
```
